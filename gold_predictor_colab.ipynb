```python
# ── Cell 3: Set your ngrok auth token ────────────────────────────────────────
#
# 1. Go to https://dashboard.ngrok.com/signup  (free account)
# 2. Copy your auth token from https://dashboard.ngrok.com/get-started/your-authtoken
# 3. Paste it below between the quotes
#
NGROK_TOKEN = "paste_your_token_here"   # <── change this

from pyngrok import ngrok, conf
conf.get_default().auth_token = NGROK_TOKEN
print('✅ ngrok configured')
```

```python
# ── Cell 4: Launch the dashboard ─────────────────────────────────────────────
import subprocess, threading, time
from pyngrok import ngrok

# Start Streamlit in the background
proc = subprocess.Popen(
    ["streamlit", "run", "app.py",
     "--server.port", "8501",
     "--server.headless", "true",
     "--server.enableCORS", "false",
     "--server.enableXsrfProtection", "false"],
    stdout=subprocess.PIPE,
    stderr=subprocess.STDOUT
)

# Give Streamlit a moment to start
time.sleep(5)

# Open a public tunnel to port 8501
public_url = ngrok.connect(8501)

print("="*55)
print(f"  🥇 Gold Predictor is LIVE!")
print(f"  👉 Open this link: {public_url}")
print("="*55)
print("\nLeave this cell running. Close the tunnel with Cell 5.")
```

```python
# ── Cell 5: Stop the app (run when done) ─────────────────────────────────────
ngrok.kill()
proc.terminate()
print('App stopped.')
```
